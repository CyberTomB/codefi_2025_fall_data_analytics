## How to Filter a DataFrame

1. Target the "series" (column) and use a comparative expression
2. Put the returned boolean column "back in" to the original dataframe, to exclude False rows
3. Profit

```python
step_one = caffeine_df['Caffeine (mg)'] > 100

step_one.head()

step_two = caffeine_df[step_one]

step_two.head() #Filtered!!
```

```python


filtered_caffeine_df = caffeine_df[caffeine_df['Caffeine (mg)'] > 100]

filtered_caffeine_df.groupby('type').agg('count')
```
